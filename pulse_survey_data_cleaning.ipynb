## Bring in data

```python
raw_data = pd.read_csv("pulse2022_puf_51.csv")
raw_data.head()
```

```python
dictionary = pd.read_csv('pulse2022_data_dictionary.csv')
print(len(dictionary))

dictionary.columns = dictionary.iloc[3,:].fillna('description_updated')
dictionary = dictionary.iloc[4:,:]
dictionary = dictionary[~dictionary['Instrument Q #'].isnull()]
dictionary.drop(['Description'], axis = 1, inplace = True)

print(len(dictionary))
dictionary.head()
```

```python
list_of_vars = ['BIRTH','HISPANIC','RACE','EEDUC','MS','GENID','SEXUAL','VACC','DOSES','BOOST','NOBSTR','COVID','TREAT','SYMPTOMS','LONGCOVID','SYMPT','WRKLOSS','ANYWORK'
,'KINDWORK','NOWRK','SETTING','UI_','EXPNS_DIF','TWDAYS','SNAP_YN','HLTHINS','SEEING','HEARING','REMEMBERING','MOBILITY','SELFCARE','UNDERSTAND','MNTHSBHIND','RENTASSIST','EVICT',
'FORCLOSE','ENERGY','BILL','INCOME','ABIRTH_YEAR','AGENID_BIRTH','AHISPANIC','ARACE','AEDUC','PRIVHLTH','PUBHLTH','REGION','SCRAM', 'PRICE', 'EXPNS_DIF', 'FOOD', 'TENURE', 
'RENT', 'MORT', 'TEMP', 'EST']
```

```python
dictionary_new = pd.DataFrame()
for var in list_of_vars:
    # print(var)
    dictionary_new = dictionary_new.append(dictionary.iloc[np.where(dictionary['Variable'].str.contains(var))])
    
print(len(dictionary_new))
```

```python
raw_data.columns[raw_data.columns.str.contains('BIRTH')].tolist()
```

```python
list_of_cols = []
for var in list_of_vars:
    list_of_cols.append(raw_data.columns[raw_data.columns.str.contains(var)].tolist())
    
def flat(lis):
    flatList = []
    # Iterate with outer list
    for element in lis:
        if type(element) is list:
            # Check if type is list than iterate through the sublist
            for item in element:
                flatList.append(item)
        else:
            flatList.append(element)
    return flatList
    
new_list = flat(list_of_cols)
new_list = list(set(new_list))
raw_data_new = raw_data.copy()[new_list]
raw_data_new.head()
```

## Create some fields for analysis

```python
### creating flags for those in marginalized groups, based on questions about gender expression, sexuality, race, ethnicity, etc.
raw_data_new['cisgender'] = np.where(raw_data_new['EGENID_BIRTH']==raw_data_new['GENID_DESCRIBE'],1,np.where(raw_data_new['GENID_DESCRIBE']<-85,np.NaN,0))
raw_data_new['lgbtq'] = np.where((1,3,4,5) in raw_data_new['SEXUAL_ORIENTATION'],1,np.where(raw_data_new['SEXUAL_ORIENTATION']<-85,np.NaN,0))
raw_data_new['hispanic'] = np.where(raw_data_new['RHISPANIC']==2,1,0)
raw_data_new['racial_minority'] = np.where(raw_data_new['RRACE']!=1,1,0)
raw_data_new['college_degree'] = np.where(raw_data_new['EEDUC']>=5,1,0)
raw_data_new['high_school_degree'] = np.where(raw_data_new['EEDUC']>=3,1,0)

### flags for other descriptors
## current fields: INCOME (1='Less than $25,000', 2='$25,000 - $34,999', 3='$35,000 - $49,999', 4='$50,000 - $74,999', 5='$75,000 - $99,999', 6'$100,000 - $149,999', 
## 7='$150,000 - $199,999', 8='$200,000 and above'), EST_ST (state), EST_MSA (metropolitan area)
raw_data_new['married'] = np.where(raw_data_new['MS']<-85,np.NaN,np.where(raw_data_new['MS']==1,1,0))
raw_data_new['own_home'] = np.where(raw_data_new['TENURE']<-85,np.NaN,np.where(raw_data_new['TENURE']<3,1,0))
raw_data_new['urban'] = np.where(raw_data_new['EST_MSA']!='',1,0)

### flags for job related attributes
## current fields: RSNNOWRKRV (1='I did not want to be employed at this time', 2='I am/was sick with coronavirus symptoms or caring for someone who was sick with coronavirus symptoms),
## KINDWORK, SETTING, UI_APPLYRV (1='Yes', 2='No'), UI_RECVRV (1='Yes', 2='No'), UI_RECVNOW (1='Yes', 2='No'), TWDAYS_RESP (1='Yes, for 1-2 days', 2='Yes, for 3-4 days', 
## 3='Yes, for 5 or more days', 4='No')
raw_data_new['unemployed'] = np.where(raw_data_new['ANYWORK']<-85,np.NaN,np.where(raw_data_new['ANYWORK']==2,1,0))
raw_data_new['recent_job_loss_past_4w'] = np.where(raw_data_new['WRKLOSSRV']<-85,np.NaN, np.where(raw_data_new['WRKLOSSRV']==1,1,0))

### flags for health-related attributes
## current fields: RECVDVACC (1='Yes', 2='No'), NUMDOSES (2='I received 2 doses of a two-shot series or a single dose vaccine like Johnson & Johnson', 
## 1='I received one dose of a two-shot series like Moderna or Pfizer'), BOOSTERRV (1='Yes', 2='No'), HADCOVIDRV (1='Yes', 2='No'), WHENCOVID (1='Within the last four weeks', 
## 2='More than four weeks ago', 3='Both'), TREATORAL (1='Yes, received oral antiviral medications (examples: Paxlovid, molnupiravir)', 
## 2='No, did not receive oral antiviral medications (examples: Paxlovid, molnupiravir)'), TREATMONO 1='Yes, received monoclonal antibody treatments (example: sotrovimab)', 
## 2='No, did not receive monoclonal antibody treatments (example: sotrovimab)'), SYMPTOMS (1='No symptoms', 2='Mild symptoms', 3='Moderate symptoms', 
## 4='Severe symptoms'), LONGCOVID (1='Yes', 2='No'), SYMPTMNOW (1='Yes', 2='No'), SYMPTMIMPCT (1='Yes, a lot', 1='Yes, a little', 3='Not at all')
raw_data_new['covid_treatment'] = np.where((raw_data_new['TREATORAL']==1)|(raw_data_new['TREATMONO']==1),1,np.where(
    (raw_data_new['TREATORAL']==2)&(raw_data_new['TREATMONO']==2),0,np.NaN))
raw_data_new['no_covid_trtmnt_reason'] = np.where((raw_data_new['RSNNOTREAT4']==1)|(raw_data_new['RSNNOTREAT6']==1)|(raw_data_new['RSNNOTREAT8']==1),"Treatments unaccessible",
np.where((raw_data_new['RSNNOTREAT7']==1)|(raw_data_new['RSNNOTREAT9']==1),"Lack of education",
np.where((raw_data_new['RSNNOTREAT1']==1)|(raw_data_new['RSNNOTREAT2']==1)|(raw_data_new['RSNNOTREAT3']==1),"Opted no",np.where(
    raw_data_new['RSNNOTREAT5']==1, "Worried about side effects","Other"))))
raw_data_new['health_insurance'] = np.where((raw_data_new['PRIVHLTH']==2)&(raw_data_new['PUBHLTH']==2),0,np.where((raw_data_new['PRIVHLTH'] == 3)|
(raw_data_new['PUBHLTH'] == 3),np.NaN,1))
raw_data_new.rename(columns = {'HLTHINS1':'employer_insurance', 'HLTHINS2':'direct_purchase_insurance', 'HLTHINS3':'medicare_inaurance', 'HLTHINS3':'medicaid_inaurance', 
'HLTHINS5':'military_insurance', 'HLTHINS6':'va_insurance', 'HLTHINS7':'indian_hs_insurance', 'HLTHINS8':'other_insurance'}, inplace = True)
raw_data_new['visual_disability'] = np.where(raw_data_new['SEEING']<-85, np.NaN, np.where(raw_data_new['SEEING']<2,0,1))
raw_data_new['hearing_disability'] = np.where(raw_data_new['HEARING']<-85, np.NaN, np.where(raw_data_new['HEARING']<2,0,1))
raw_data_new['memory_concentration_disability'] = np.where(raw_data_new['REMEMBERING']<-85, np.NaN, np.where(raw_data_new['REMEMBERING']<2,0,1))
raw_data_new['mobility_disability'] = np.where(raw_data_new['MOBILITY']<-85, np.NaN, np.where(raw_data_new['MOBILITY']<2,0,1))
raw_data_new['selfcare_disability'] = np.where(raw_data_new['SELFCARE']<-85, np.NaN, np.where(raw_data_new['SELFCARE']<2,0,1))
raw_data_new['communication_disability'] = np.where(raw_data_new['UNDERSTAND']<-85, np.NaN, np.where(raw_data_new['UNDERSTAND']<2,0,1))

### flags for financial stress
## current fields: PRICESTRESS (1='Very stressful', 2='Moderately stressful', 3='A little stressful', 4='Not at all stressful'), EXPNS_DIF (1='Not at all difficult', 
## 2='A little difficult', 3='Somewhat difficult', 4'Very difficult'), SNAPYN (1='Yes', 2='No'), TMNTHSBHND (months behind in rent), RENTASSIST 
## (1='My household applied and received assistance', 2='My household applied and is waiting for a response', 3='My household applied and the application was denied', 
## 4='My household did not apply'), EVICT (1='Very likely', 2='Somewhat likely', 3='Not very likely', 4='Not likely at all'), FORCLOSE 
## (1='Very likely', 2='Somewhat likely', 3='Not very likely', 4='Not likely at all')
raw_data_new['less_spending'] = np.where((raw_data_new['PRICECOPE1']==1)|(raw_data_new['PRICECOPE2']==1)|(raw_data_new['PRICECOPE3']==1)|(raw_data_new['PRICECOPE4']==1)|
(raw_data_new['PRICECOPE5']==1)|(raw_data_new['PRICECOPE6']==1)|(raw_data_new['PRICECOPE7']==1)|(raw_data_new['PRICECOPE8']==1)|(raw_data_new['PRICECOPE9']==1)
|(raw_data_new['PRICECOPE13']==1)|(raw_data_new['PRICECOPE14']==1)|(raw_data_new['PRICECOPE16']==1),1,0)
raw_data_new['more_borrowing'] = np.where((raw_data_new['PRICECOPE12']==1)|(raw_data_new['PRICECOPE15']==1)|(raw_data_new['PRICECOPE17']==1),1,0)
raw_data_new['less_saving'] = np.where(raw_data_new['PRICECOPE11']==1,1,0)
raw_data_new['more_income'] = np.where(raw_data_new['PRICECOPE10']==1,1,0)
raw_data_new['food_insecure_past_7d'] = np.where(raw_data_new['CURFOODSUF']<-85, np.NaN, np.where(raw_data_new['CURFOODSUF']<2,0,1))
raw_data_new['couldnt_afford_food'] = np.where(raw_data_new['FOODRSNRV1']==1,1,0)
raw_data_new['couldnt_access_food'] = np.where((raw_data_new['FOODRSNRV2']==1)|(raw_data_new['FOODRSNRV3']==1),1,0)
raw_data_new['had_free_meal_past_7d'] = np.where(raw_data_new['FREEFOOD']<-85,np.NaN, np.where(raw_data_new['FREEFOOD']<2, 1,0))
raw_data_new['home_payments_late'] = np.where((raw_data_new['RENTCUR']<-85)&(raw_data_new['MORTCUR']<-85),np.NaN, 
np.where((raw_data_new['RENTCUR']==2)|(raw_data_new['MORTCUR']==2),1,0))
raw_data_new['likely_eviction'] = np.where(raw_data_new['EVICT']<-85,np.NaN, np.where(raw_data_new['EVICT']<3, 1,0))
raw_data_new['likely_forclosure'] = np.where(raw_data_new['FORCLOSE']<-85,np.NaN, np.where(raw_data_new['FORCLOSE']<3, 1,0))
raw_data_new['trouble_paying_energy'] = np.where((raw_data_new['ENERGY']<-85)&(raw_data_new['HSE_TEMP']<-85)&(raw_data_new['ENRGY_BILL']<-85),np.NaN, 
np.where((raw_data_new['ENERGY']<4)|(raw_data_new['HSE_TEMP']<4)|(raw_data_new['ENRGY_BILL']<4),1,0))
```
